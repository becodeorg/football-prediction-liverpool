```python
# 1ï¸âƒ£ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2ï¸âƒ£ Load the dataset
file_path = r"C:\Users\User\OneDrive - Arc-en-Ciel\Desktop\last foot\dataset.csv"  # Change to your CSV path
df = pd.read_csv(file_path)

# 3ï¸âƒ£ Basic dataset overview
print("âœ… Dataset shape:", df.shape)
print("\nðŸ“Œ First 5 rows:")
display(df.head())

print("\nðŸ“Œ Columns info:")
df.info()

print("\nðŸ“Œ Statistical summary for numeric columns:")
display(df.describe())

# 4ï¸âƒ£ Missing values analysis
print("\nðŸ“Œ Missing values per column:")
missing_values = df.isnull().sum().sort_values(ascending=False)
display(missing_values)

# Visualize missing values
plt.figure(figsize=(10,4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()

# 5ï¸âƒ£ Target column distribution (if exists)
if 'FTR' in df.columns:
    print("\nðŸ“Š Match Results Distribution:")
    df['FTR'].value_counts().plot(kind='bar', title='Match Results Distribution', figsize=(5,3))
    plt.xlabel('Result')
    plt.ylabel('Count')
    plt.show()

# 6ï¸âƒ£ Correlation heatmap for numeric columns
if df.select_dtypes(include='number').shape[1] > 1:
    plt.figure(figsize=(10,6))
    sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
    plt.title("Correlation Heatmap (Numeric Columns)")
    plt.show()
```

```python

# 1ï¸âƒ£ Calculate the percentage of missing values per column
missing_percent = df.isnull().mean() * 100
missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)
print("Missing values percentage per column:")
print(missing_percent)

# 2ï¸âƒ£ Step 1: Fill columns with a small percentage of missing values (<5%) using the median
small_missing_cols = missing_percent[missing_percent < 5].index
for col in small_missing_cols:
    df[col].fillna(df[col].median(), inplace=True)

print(f"\nâœ… Filled {len(small_missing_cols)} columns with median (small missing %).")

# 3ï¸âƒ£ Step 2: Handle betting columns dynamically
betting_keywords = ['WH', 'BW', 'B365']
betting_cols = [col for col in df.columns if any(key in col for key in betting_keywords)]

# Function to find alternative columns for a given betting column
def find_alternatives(col_name):
    # Use the same suffix across other betting providers
    suffix = col_name[2:]  # remove WH/BW prefix
    return [alt for alt in betting_cols if alt != col_name and alt.endswith(suffix)]

# Fill missing values in betting columns
for col in betting_cols:
    if df[col].isnull().sum() > 0:
        # Try to fill from alternative columns first
        for alt in find_alternatives(col):
            df[col].fillna(df[alt], inplace=True)
        # Fill any remaining missing values with the median
        df[col].fillna(df[col].median(), inplace=True)

print("\nâœ… Betting columns filled using smart alternatives + median.")

# 4ï¸âƒ£ Final check for remaining missing values
remaining_missing = df.isnull().sum()[df.isnull().sum() > 0]
if remaining_missing.empty:
    print("\nðŸŽ‰ All missing values handled successfully!")
else:
    print("\nâš ï¸ Remaining missing values:")
    print(remaining_missing)
```
